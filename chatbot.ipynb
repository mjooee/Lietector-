```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm
import urllib.request
from sentence_transformers import SentenceTransformer
```

```python
#챗봇 QA 엑셀 파일 불러오기
train_data = pd.read_excel('/content/drive/MyDrive/대처방법_챗봇.xlsx', sheet_name = '챗봇 QA')
train_data.head()
```

```python
#코사인 유사도
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
help_text = "문의하신 사항을 제대로 이해하지 못했습니다. 버튼으로 선택해주시거나 다시 문의해주십시오."
```

한국어 버전의 버트 https://github.com/jhgan00/ko-sentence-transformers

```python
model_original = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
model_finetune_10 = SentenceTransformer('/content/drive/MyDrive/chatbot_10')
model_finetune_5 = SentenceTransformer('/content/drive/MyDrive/chatbot_5')
```

```python
train_data['embedding_o'] = train_data.apply(lambda row: model_original.encode(row.text), axis = 1)
train_data['embedding_f10'] = train_data.apply(lambda row: model_finetune_10.encode(row.text), axis = 1)
train_data['embedding_f5'] = train_data.apply(lambda row: model_finetune_5.encode(row.text), axis = 1)
```

```python
#원래 모델은 전체적으로 유사도가 높게 나오고 파인튜닝 모델은 전체적으로 유사도가 낮게 나와서 각각 0.7, 0.6으로 설정
#입력 질문과 가장 유사한 질문 데이터를 찾아서 해당 질문데이터의 대답데이터 반환
def return_answer_o(question): #오리지널
    embedding = model_original.encode(question)
    train_data['score_o'] = train_data.apply(lambda x: cos_sim(x['embedding_o'], embedding), axis=1)
    print("유사도:", train_data.loc[train_data['score_o'].idxmax()]['score_o'])
    if train_data.loc[train_data['score_o'].idxmax()]['score_o'] < 0.7:
      return help_text
    print("매칭된 질문은: ", train_data.loc[train_data['score_o'].idxmax()]['text'])
    return train_data.loc[train_data['score_o'].idxmax()]['answer']

def return_answer_f10(question): #파인튜닝
    embedding = model_finetune_10.encode(question)
    train_data['score_f10'] = train_data.apply(lambda x: cos_sim(x['embedding_f10'], embedding), axis=1)
    print("유사도:", train_data.loc[train_data['score_f10'].idxmax()]['score_f10'])
    if train_data.loc[train_data['score_f10'].idxmax()]['score_f10'] < 0.65:
      return help_text
    print("매칭된 질문은: ", train_data.loc[train_data['score_f10'].idxmax()]['text'])
    return train_data.loc[train_data['score_f10'].idxmax()]['answer']

def return_answer_f5(question): #파인튜닝
    embedding = model_finetune_5.encode(question)
    train_data['score_f5'] = train_data.apply(lambda x: cos_sim(x['embedding_f5'], embedding), axis=1)
    print("유사도:", train_data.loc[train_data['score_f5'].idxmax()]['score_f5'])
    if train_data.loc[train_data['score_f5'].idxmax()]['score_f5'] < 0.65:
      return help_text
    print("매칭된 질문은: ", train_data.loc[train_data['score_f5'].idxmax()]['text'])
    return train_data.loc[train_data['score_f5'].idxmax()]['answer']
```
